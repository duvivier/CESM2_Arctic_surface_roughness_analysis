```python
import xarray as xr
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.path as mpath
from matplotlib.gridspec import GridSpec
from scipy.stats import linregress,pearsonr, t, ttest_ind
```

## Calculate theoretical roughnesses

```python
# calculate the wind speed dependent ocean drag coefficient
# u --> wind speed
# rdn_ocn = sqrt((0.0027/u)+0.000142+(0.0000764*u))
# cdn_ocn = rdn_ocn * rdn_ocn

# create wind speed array
u = np.arange(0,25,0.25)
# set first value to not be 0 or we get undefined
u[0] = 0.0001

# calculate drag
rdn_ocn = np.sqrt((0.0027/u)+0.000142+(0.0000764*u))
cdn_ocn = rdn_ocn * rdn_ocn
```

```python
# calculate the (constant) sea ice drag coefficients used at all wind speeds
# k = 0.4 --> von karman constant 
# zref = 10m --> reference height
# iceruf = 0.0005m --> roughness length over ice
# rdn_ice = k/ln(zref/iceruf)
# cdn_ice = rdn_ice * rdn_ice

# set constants
k = 0.4
zref = 10
iceruf = 0.0005

# calculate drag
rdn_ice = k/np.log(zref/iceruf)
cdn_ice = rdn_ice * rdn_ice

# make it an array
cdn_ice_array = np.full(u.shape,cdn_ice)
```

## Plot the figure

```python
# multiply each array by 1000 to plot
cdn_ice_plt = cdn_ice_array * 1000
cdn_ocn_plt = cdn_ocn * 1000
```

```python
# create figure
fig = plt.figure(figsize=(20,10))
fout = 'fig_2_drag_coefficients'

# Make subplot - note it's nrow x ncol x index (starting upper left)
ax = fig.add_subplot(1,1,1)
# plot Cdn_ice
ax.plot(u,cdn_ice_plt,label='Cdn_ice - Eq.1',color='blue',linestyle='-',linewidth=3)
# plot Cdn_ocn
ax.plot(u,cdn_ocn_plt,label='Cdn_ocn - Eq.2',color='black',linestyle='-',linewidth=3)

# add stuff to plot
plt.xlabel('wind speed (m/s)',fontsize=30)
plt.xticks(fontsize=30)
plt.xlim([0,25])
plt.ylabel('1000*Cdn',fontsize=30)
plt.yticks(fontsize=30)
plt.ylim([0,3])
plt.legend(loc='lower right', fontsize=30, ncol=1);
    
# Finalize figure and save
plt.title('Theoretical Drag Coefficients' ,fontsize=30)
fig = plt.savefig(fout+'.png', bbox_inches='tight', dpi=200)
```
