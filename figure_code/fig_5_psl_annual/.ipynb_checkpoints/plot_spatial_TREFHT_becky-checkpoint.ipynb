## Read the CESM-LE data 

We will use [`intake-esm`](https://intake-esm.readthedocs.io/en/latest/), which is a data catalog tool.
It enables querying a database for the files we want, then loading those directly as an `xarray.Dataset`.

First step is to set the "collection" for the CESM-LE, which depends on a json file conforming to the [ESM Catalog Specification](https://github.com/NCAR/esm-collection-spec).

```python
catalog_file = '/glade/collections/cmip/catalog/intake-esm-datastore/catalogs/glade-cesm2-le.json'

cat = intake.open_esm_datastore(catalog_file)
```

```python
forcing = 'cmip6'  # do not want smbb data
freq = 'month_1'

subset_1 = cat.search(variable=var_in_1, forcing_variant=forcing, frequency=freq )
```

```python
# check that we only have cmip6, not smbb, data
member_id = list(subset_1.df.experiment.unique())
print(member_id)
```

```python
%%time
with dask.config.set(**{'array.slicing.split_large_chunks': True}):
    dsets_1 = subset_1.to_dataset_dict(cdf_kwargs={'chunks': {'time':240}, 'decode_times': True})
```

```python
# load in the future datasets
historicals = []
futures = []

for key in sorted(dsets_1.keys()):
    if 'historical' in key:
        historicals.append(dsets_1[key])
        print(key)
    elif 'ssp370' in key:
        futures.append(dsets_1[key])
        print(key)
```

```python
# Now put these into an array
historical_ds = xr.concat(historicals, dim='member_id')
future_ds = xr.concat(futures, dim='member_id')
```

```python
# note that the historical and future xarray datasets have the same coordinates and dimensions *except* time, 
# so we need to concatenate over time
ds_all = xr.concat([historical_ds,future_ds],dim='time')
```

```python
ds_all.time
```

```python
# Shift months by one to be center of time period.
# Take average of the time bounds to get middle of month
# will lose some attributes with time, so may need to put this back in later...
ds_all['time'] = ds_all.time_bnds.load().mean(dim='nbnd').sel(member_id='r1i1281p1f1')
```

```python
t2m_le = ds_all[var_in_1]
```

```python
t2m_le.persist()
```

```python
t2m_le.sel(member_id='r1i1281p1f1').isel(time=100).plot()
```

## Convert to C

```python
t2m_le = t2m_le - 273.15
```

```python
t2m_le.sel(member_id='r1i1281p1f1').isel(time=0).plot()
```

## Calculate ensemble means

```python
t2m_le_ens_mean = t2m_le.mean(dim='member_id')
```

```python
t2m_le_ens_mean
```

## Calculate global and Arctic means

```python
t2m_global = t2m_le_ens_mean.mean(dim=['lat','lon'])
```

```python
t2m_global
```

```python
t2m_arctic = t2m_le_ens_mean.where(t2m_le_ens_mean.lat >= 70, drop=True).mean(dim=['lat','lon'])
```

```python
t2m_arctic
```

## Load the data to make the annual mean calculation faster

```python
t2m_global.load()
t2m_arctic.load()
```

## Calculate annual means

```python
# find total years
xarr_le = t2m_le.coords['time.year'][(t2m_le.coords['time.month']==1)]
```

```python
# Loop through seasons - CESM2-LE

# make numpy array to fill and specify dimensions we want
array_le_global = np.zeros([len(xarr_le)])
array_le_arctic = np.zeros([len(xarr_le)])

# now loop through years to get the seasonal average by year for each ensemble member
for yy in xarr_le:
    if yy == 1850:
        y_count = 0
    else: 
        y_count = y_count+1 
    ### select only the indexes for this year
    # global
    temp1 = t2m_global.isel(time=t2m_global.time.dt.year.isin([yy]))
    temp2 = temp1.mean(dim='time')
    array_le_global[y_count] = temp2 
    # arctic
    temp1 = t2m_arctic.isel(time=t2m_arctic.time.dt.year.isin([yy]))
    temp2 = temp1.mean(dim='time')
    array_le_arctic[y_count] = temp2  
```

```python
print(array_le_global.shape)
print(array_le_arctic.shape)
```

```python
# convert the numpy array to a xarray for easier plotting
t2m_ann_global = xr.DataArray(array_le_global,dims=('time'))
t2m_ann_arctic = xr.DataArray(array_le_arctic,dims=('time'))
```

```python
# set coordinate arrays
t2m_ann_global['time'] = xarr_le
t2m_ann_arctic['time'] = xarr_le
```

```python
t2m_ann_global.plot()
```

```python
t2m_ann_arctic.plot()
```

## Calculate anomalies from 1850s

```python
t2m_1850s_global = t2m_ann_global.where(t2m_ann_global.time < 1860, drop=True).mean(dim='time')
t2m_1850s_arctic = t2m_ann_arctic.where(t2m_ann_arctic.time < 1860, drop=True).mean(dim='time')
```

```python
t2m_anom_global = t2m_ann_global - t2m_1850s_global
t2m_anom_arctic = t2m_ann_arctic - t2m_1850s_arctic
```

```python
t2m_anom_global.plot()
```

```python
t2m_anom_arctic.plot()
```

## Export data

```python
# make array of the global data - raw and anomaly
array_le_global = np.zeros([len(xarr_le),3])

### fill array
# years
array_le_global[:,0] = xarr_le.values
# raw data
array_le_global[:,1] = t2m_ann_global.values
# anomaly from 1850s
array_le_global[:,2] = t2m_anom_global.values

# save file
np.savetxt('t2m_global.txt',array_le_global)
```

```python
# make array of the global data - raw and anomaly
array_le_arctic = np.zeros([len(xarr_le),3])

### fill array
# years
array_le_arctic[:,0] = xarr_le.values
# raw data
array_le_arctic[:,1] = t2m_ann_arctic.values
# anomaly from 1850s
array_le_arctic[:,2] = t2m_anom_arctic.values

# save file
np.savetxt('t2m_arctic.txt',array_le_arctic)
```
