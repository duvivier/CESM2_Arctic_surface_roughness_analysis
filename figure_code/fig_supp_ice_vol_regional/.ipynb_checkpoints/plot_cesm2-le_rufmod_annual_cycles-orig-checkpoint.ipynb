## Read in the processed data

This reads in pre-processed regional total ice area or volume for the Arctic.

Total Ice area = sum over lat/lon of (aice*tarea). Units = km2

Ice volume = sum over lat/lon of (hi*tarea). Units = km3


```python
# list the variables to load
var_in = 'hi'
```

```python
# set base directory where all data live
data_dir = '/glade/p/cgd/ppc/duvivier/cesm2_arctic_cyclones/DATA/regional_timeseries/'

# set file names
fin_le = 'CESM2-LE_all_transient_regional_total_'+var_in
fin_rufmod = 'rufmod_all_transient_regional_total_'+var_in
```

```python
ds_le = xr.open_dataset(data_dir+fin_le+'.nc',decode_times=True)
ds_rufmod = xr.open_dataset(data_dir+fin_rufmod+'.nc',decode_times=True)
```

```python
# double check that year/month is not shifted incorrectly
ds_le.coords['time.month']
ds_rufmod.coords['time.month']
```

```python
# check what regions we have totals for
ds_le.region

# should match...
mask_in = ['nh_mask', 'Lab_mask', 'GIN_mask', 'Bar_mask', 'ArcOc_mask', 'Sib_mask', 'Beau_mask', 'Bering_mask', 'Hudson_mask', 'CAArch_mask', 'Okhotsk_mask']
```

```python
print(ds_le.units)
print(ds_le.longname)

total_le = ds_le['regional_totals']
total_rufmod = ds_rufmod['regional_totals']
```

```python
total_le
```

### Calculate the total ice area or volume in the Arctic Ocean + Barents Sea ONLY

```python
reg1 = 'ArcOc_mask'
reg2 = 'Bar_mask'
reg = reg1+'_and_'+reg2
```

```python
data_le = total_le.sel(region=reg1) + total_le.sel(region=reg2)
data_rufmod = total_rufmod.sel(region=reg1) + total_rufmod.sel(region=reg2)
```

### Calculate the monthly means by decade

```python
# make lists of both the months and decades we'll use
decades = list(range(1850,2100,10))
months = list(range(1,13,1))
```

```python
#specify some arrays that will become coordinate arrays
month_names = ['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun', 'Jul', 'Aug', 'Sep', 'Oct', 'Nov', 'Dec']
decade_names = ['1850', '1860', '1870', '1880', '1890','1900','1910','1920','1930',
                '1940','1950','1960','1970','1980','1990','2000','2010','2020','2030',
                '2040','2050','2060','2070','2080','2090']
```

#### CESM2-LE calculations

```python
# For CESM2-LE: loop through months and decades to get all the annual cycles

# make numpy array to fill and specify dimensions we want
array_le = np.zeros([len(total_le.member_id),len(months),len(decades)])
#array_le.shape

# decade loop
for dec in decades:
    print('Calculating monthly means for '+str(dec))
    # need to specify which spot in the array we'll fill for the decade
    if dec == 1850:
        count = 0
    else: 
        count = count+1
    # set the start and end years for this decade
    yy_st = dec
    yy_ed = yy_st + 10
    yy = list(range(yy_st,yy_ed,1))
    # subset the data for these years only but keep all months
    temp = data_le.isel(time=data_le.time.dt.year.isin([yy]))
    
    # month loop
    for mm in months:
        #print('Finding month '+str(mm))
        temp2 = temp.isel(time=temp.time.dt.month.isin([mm]))
        array_le[:,(mm-1),count] = temp2.mean(dim='time')
```

```python
# check the shape
print(array_le.shape)    
```

```python
# convert the numpy array to a xarray
decadal_means_le = xr.DataArray(array_le,dims=('member_id','month','decade'))
```

```python
# assign some coordinate arrays
decadal_means_le['member_id'] = total_le.member_id
decadal_means_le['decade'] = np.array(decades)
decadal_means_le['month'] = np.array(months)
```

```python
decadal_means_le.decade
```

#### Rufmod experiments

```python
# For Rufmod: loop through months and decades to get all the annual cycles
# NOTE that we had to change the loop for decades to only do 1950 onward as there isn't data before then for this expt.

# make numpy array to fill and specify dimensions we want
array_rufmod = np.zeros([len(total_rufmod.member_id),len(months),len(decades[10:])])
#array_le.shape

# decade loop
for dec in decades[10:]:
    print('Calculating monthly means for '+str(dec))
    # need to specify which spot in the array we'll fill for the decade
    if dec == 1950:
        count = 0
    else: 
        count = count+1
    # set the start and end years for this decade
    yy_st = dec
    yy_ed = yy_st + 10
    yy = list(range(yy_st,yy_ed,1))
    # subset the data for these years only but keep all months
    temp = data_rufmod.isel(time=data_rufmod.time.dt.year.isin([yy]))
    
    # month loop
    for mm in months:
        #print('Finding month '+str(mm))
        temp2 = temp.isel(time=temp.time.dt.month.isin([mm]))
        array_rufmod[:,(mm-1),count] = temp2.mean(dim='time')
```

```python
# check the shape
print(array_rufmod.shape)    
```

```python
# convert the numpy array to a xarray
decadal_means_rufmod = xr.DataArray(array_rufmod,dims=('member_id','month','decade'))
```

```python
# assign some coordinate arrays
decadal_means_rufmod['member_id'] = total_rufmod.member_id
decadal_means_rufmod['decade'] = np.array(decades[10:])
decadal_means_rufmod['month'] = np.array(months)
```

```python
decadal_means_rufmod.decade
```

### Plotting!

```python
# Plot rough version just to check
import nc_time_axis
# Setup the figure
fig = plt.figure(figsize=(12, 8))
ax = plt.subplot(111)

#CESM2-LE
# Loop through and plot each member_id all months in 251 years
for member_id in decadal_means_le.member_id.values:
    decadal_means_le.sel(decade=1980,member_id=member_id).plot(ax=ax)
#rufmod
# Loop through and plot each member_id all months overlay in black
for member_id in decadal_means_rufmod.member_id.values:
    decadal_means_rufmod.sel(decade=1980,member_id=member_id).plot(ax=ax,color='black')
```

```python
#set some labeling info for plots
if var_in == 'aice':
    out_tag = 'ice_area'
    label = 'Ice Area'
    # convert to 1000000's of km3
    decadal_means_le = decadal_means_le/(1000000.0)
    decadal_means_rufmod = decadal_means_rufmod/(1000000.0)
    units = '($10^{6}$ $km^{2}$)'
else:
    out_tag = 'ice_vol'
    label = 'Ice Volume'
    # convert to 1000's of km3
    decadal_means_le = decadal_means_le/(1000.0)
    decadal_means_rufmod = decadal_means_rufmod/(1000.0)
    units = '($10^{3}$ $km^{3}$)'
```

```python
# calculate means and statistical significance by decades
sigval1 = 0.05
sigval2 = 0.1

# Choose every other decade
dec = 2020
temp1 = decadal_means_le.sel(decade=dec)
temp2 = decadal_means_rufmod.sel(decade=dec)
tval,pval = ttest_ind(temp1,temp2,axis=0,equal_var=False)
le_data_2020 = temp1.mean(dim='member_id')
rufmod_data_2020 = temp2.mean(dim='member_id')
le_data_2020_sig1 = le_data_2020.where(pval <= sigval1)
rufmod_data_2020_sig1 = rufmod_data_2020.where(pval <= sigval1)
le_data_2020_sig2 = le_data_2020.where(pval <= sigval2)
rufmod_data_2020_sig2 = rufmod_data_2020.where(pval <= sigval2)

dec = 2040
temp1 = decadal_means_le.sel(decade=dec)
temp2 = decadal_means_rufmod.sel(decade=dec)
tval,pval = ttest_ind(temp1,temp2,axis=0,equal_var=False)
le_data_2040 = temp1.mean(dim='member_id')
rufmod_data_2040 = temp2.mean(dim='member_id')
le_data_2040_sig1 = le_data_2040.where(pval <= sigval1)
rufmod_data_2040_sig1 = rufmod_data_2040.where(pval <= sigval1)
le_data_2040_sig2 = le_data_2040.where(pval <= sigval2)
rufmod_data_2040_sig2 = rufmod_data_2040.where(pval <= sigval2)

dec = 2060
temp1 = decadal_means_le.sel(decade=dec)
temp2 = decadal_means_rufmod.sel(decade=dec)
tval,pval = ttest_ind(temp1,temp2,axis=0,equal_var=False)
le_data_2060 = temp1.mean(dim='member_id')
rufmod_data_2060 = temp2.mean(dim='member_id')
le_data_2060_sig1 = le_data_2060.where(pval <= sigval1)
rufmod_data_2060_sig1 = rufmod_data_2060.where(pval <= sigval1)
le_data_2060_sig2 = le_data_2060.where(pval <= sigval2)
rufmod_data_2060_sig2 = rufmod_data_2060.where(pval <= sigval2)

dec = 2080
temp1 = decadal_means_le.sel(decade=dec)
temp2 = decadal_means_rufmod.sel(decade=dec)
tval,pval = ttest_ind(temp1,temp2,axis=0,equal_var=False)
le_data_2080 = temp1.mean(dim='member_id')
rufmod_data_2080 = temp2.mean(dim='member_id')
le_data_2080_sig1 = le_data_2080.where(pval <= sigval1)
rufmod_data_2080_sig1 = rufmod_data_2080.where(pval <= sigval1)
le_data_2080_sig2 = le_data_2080.where(pval <= sigval2)
rufmod_data_2080_sig2 = rufmod_data_2080.where(pval <= sigval2)
```

```python
# set x arrays against which to plot
xarr = months
```

```python
# Plot selected regional ice area with select decades
print('Plotting decadal annual cycles for region '+reg)

# create figure
fig = plt.figure(figsize=(20,10))
fout = 'total_cesm2-le_and_rufmod_'+out_tag+'_'+reg+'_annual_cycle_transient'

# now plot that data including significance
ax = fig.add_subplot(1,1,1)
# Plot data all
ax.plot(xarr,le_data_2020,label="CESM2-LE - 2020's",color='darkblue',linestyle='-', linewidth=3, marker='o', markersize=5)
ax.plot(xarr,le_data_2040,label="CESM2-LE - 2040's",color='blue',linestyle='-', linewidth=3, marker='o', markersize=5)
ax.plot(xarr,le_data_2060,label="CESM2-LE - 2060's",color='cornflowerblue',linestyle='-', linewidth=3, marker='o', markersize=5)
ax.plot(xarr,le_data_2080,label="CESM2-LE - 2080's",color='lightskyblue',linestyle='-', linewidth=3, marker='o', markersize=5)
ax.plot(xarr,rufmod_data_2020,label="SMOOTH - 2020's",color='darkblue',linestyle=':', linewidth=3, marker='o', markersize=5)
ax.plot(xarr,rufmod_data_2040,label="SMOOTH - 2040's",color='blue',linestyle=':', linewidth=3, marker='o', markersize=5)
ax.plot(xarr,rufmod_data_2060,label="SMOOTH - 2060's",color='cornflowerblue',linestyle=':', linewidth=3, marker='o', markersize=5)
ax.plot(xarr,rufmod_data_2080,label="SMOOTH - 2080's",color='lightskyblue',linestyle=':', linewidth=3, marker='o', markersize=5)

# Plot the significant differences (90%) as bigger marker
ax.plot(xarr,le_data_2020_sig2,color='darkblue',linestyle=' ', marker='o',markersize=20,fillstyle='none')
ax.plot(xarr,le_data_2040_sig2,color='blue', linestyle=' ', marker='o',markersize=20,fillstyle='none')
ax.plot(xarr,le_data_2060_sig2,color='cornflowerblue',linestyle=' ', marker='o', markersize=20,fillstyle='none')
ax.plot(xarr,le_data_2080_sig2,color='lightskyblue',linestyle=' ',marker='o', markersize=20,fillstyle='none')
ax.plot(xarr,rufmod_data_2020_sig2,color='darkblue',linestyle=' ', marker='o', markersize=20,fillstyle='none')
ax.plot(xarr,rufmod_data_2040_sig2,color='blue',linestyle=' ', marker='o', markersize=20,fillstyle='none')
ax.plot(xarr,rufmod_data_2060_sig2,color='cornflowerblue',linestyle=' ', marker='o', markersize=20,fillstyle='none')
ax.plot(xarr,rufmod_data_2080_sig2,color='lightskyblue',linestyle=' ', marker='o', markersize=20,fillstyle='none')

# Plot the significant differences (95%) as bigger marker
ax.plot(xarr,le_data_2020_sig1,color='darkblue',linestyle=' ', marker='o',markersize=20)
ax.plot(xarr,le_data_2040_sig1,color='blue', linestyle=' ', marker='o',markersize=20)
ax.plot(xarr,le_data_2060_sig1,color='cornflowerblue',linestyle=' ', marker='o', markersize=20)
ax.plot(xarr,le_data_2080_sig1,color='lightskyblue',linestyle=' ',marker='o', markersize=20)
ax.plot(xarr,rufmod_data_2020_sig1,color='darkblue',linestyle=' ', marker='o', markersize=20)
ax.plot(xarr,rufmod_data_2040_sig1,color='blue',linestyle=' ', marker='o', markersize=20)
ax.plot(xarr,rufmod_data_2060_sig1,color='cornflowerblue',linestyle=' ', marker='o', markersize=20)
ax.plot(xarr,rufmod_data_2080_sig1,color='lightskyblue',linestyle=' ', marker='o', markersize=20)

# set some labeling info
plt.xlabel('Month',fontsize=20)
plt.xticks(xarr,fontsize=20)
plt.xlim([0.5,12.5])
plt.ylabel(units,fontsize=20)
plt.yticks(fontsize=20)
plt.legend(ncol=2,loc=3,fontsize=20)

# Finalize figure and save
fig.suptitle('Total Arctic Ocean '+label+' Monthly Ensemble Mean by Decade',fontsize=20, y=0.92)  
fig.subplots_adjust(hspace=0.3)
fig = plt.savefig(fout+'.png', bbox_inches='tight', dpi=200)
```
