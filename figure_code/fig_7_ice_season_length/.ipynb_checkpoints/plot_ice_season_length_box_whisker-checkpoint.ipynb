## Process data into decadal groups to do box plot

```python
#specify some arrays that will become coordinate arrays
decades = list(range(2020,2100,10))
decade_names = ['2020','2030','2040','2050','2060','2070','2080','2090']
```

### Calculate ensemble mean ice free days per decade

- Average by decade and over ensembles to get ensemble mean of ice covered days

```python
# Loop through seasons
print('Calculating Decadal and Ensemble Means ')

# make numpy array to fill and specify dimensions we want
dec_array_le = np.zeros([len(decade_names),len(data_le.nj),len(data_le.ni)])
dec_array_smooth = np.zeros([len(decade_names),len(data_smooth.nj),len(data_smooth.ni)])

# decade loop
for dec in decades:
    print('Calculating decadal means for '+str(dec)+'s')
    # need to specify which spot in the array we'll fill for the decade
    if dec == 2020:
        count = 0
    else: 
        count = count+1
    # set the start and end years for this decade
    yy_st = dec
    yy_ed = yy_st + 10
    yy = list(range(yy_st,yy_ed,1))
    # subset the data for these years only 
    temp1 = data_le_masked.isel(time=data_le_masked.time.isin([yy]))
    temp2 = data_smooth_masked.isel(time=data_smooth_masked.time.isin([yy]))
    # stack so we can average over years and ensembles at once
    temp1a = temp1.stack(all_times=("time","member_id"))
    dec_array_le[count,:,:] = temp1a.mean(dim='all_times')
    temp2a = temp2.stack(all_times=("time","member_id"))
    dec_array_smooth[count,:,:] = temp2a.mean(dim='all_times')        
```

```python
print(dec_array_smooth.shape)
print(dec_array_le.shape)
```

```python
# convert the numpy array to a xarray for easier plotting
dec_ndays_smooth_ens_mean = xr.DataArray(dec_array_smooth,dims=('decade','nj','ni'))
dec_ndays_le_ens_mean = xr.DataArray(dec_array_le,dims=('decade','nj','ni'))
```

```python
# set coordinate arrays
dec_ndays_smooth_ens_mean['decade'] = decade_names
dec_ndays_smooth_ens_mean['nj'] = data_smooth['nj'].values
dec_ndays_smooth_ens_mean['ni'] = data_smooth['ni'].values

dec_ndays_le_ens_mean['decade'] = decade_names
dec_ndays_le_ens_mean['nj'] = data_le['nj'].values
dec_ndays_le_ens_mean['ni'] = data_le['ni'].values
```

```python
# stack into a single set of points per decade to plot
dec_ndays_smooth_ens_mean = dec_ndays_smooth_ens_mean.stack(all=("nj","ni"))
dec_ndays_le_ens_mean = dec_ndays_le_ens_mean.stack(all=("nj","ni"))
# print size to check
print(dec_ndays_smooth_ens_mean.shape)
print(dec_ndays_le_ens_mean.shape)
```

```python
# transpose so that it will plot properly
dec_ndays_smooth_ens_mean = dec_ndays_smooth_ens_mean.transpose()
dec_ndays_le_ens_mean = dec_ndays_le_ens_mean.transpose()
# print size to check
print(dec_ndays_smooth_ens_mean.shape)
print(dec_ndays_le_ens_mean.shape)
```

### Make arrays by decade for each ensemble member

```python
# get total points that are going to be in box plot for all ensembles
npts_tot_smooth = len(data_smooth.member_id)*len(data_smooth.ni)*len(data_smooth.nj)*10
npts_tot_le = len(data_le.member_id)*len(data_le.ni)*len(data_le.nj)*10

print(npts_tot_smooth)
print(npts_tot_le)
```

```python
# Loop through seasons - all ensembles
print('Calculating Decadal Values ')

# make numpy array to fill and specify dimensions we want
dec_array_le = np.zeros([len(decade_names),npts_tot_le])
dec_array_smooth = np.zeros([len(decade_names),npts_tot_smooth])

# decade loop
for dec in decades:
    print('Getting decadal groups for '+str(dec)+'s')
    # need to specify which spot in the array we'll fill for the decade
    if dec == 2020:
        count = 0
    else: 
        count = count+1
    # set the start and end years for this decade
    yy_st = dec
    yy_ed = yy_st + 10
    yy = list(range(yy_st,yy_ed,1))
    # subset the data for these years only, but all ensembles
    temp1 = data_le_masked.isel(time=data_le_masked.time.isin([yy]))
    temp2 = data_smooth_masked.isel(time=data_smooth_masked.time.isin([yy]))
    # now stack all values and put them into the correct array
    dec_array_le[count,:] = temp1.stack(all=("member_id","time","nj","ni"))
    dec_array_smooth[count,:] = temp2.stack(all=("member_id","time","nj","ni"))    
```

```python
print(dec_array_smooth.shape)
print(dec_array_le.shape)
```

```python
# convert the numpy array to a xarray for easier plotting
dec_ndays_smooth_all = xr.DataArray(dec_array_smooth,dims=('decade','all'))
dec_ndays_le_all = xr.DataArray(dec_array_le,dims=('decade','all'))
```

```python
# set coordinate arrays
dec_ndays_smooth_all['decade'] = decade_names
dec_ndays_le_all['decade'] = decade_names
```

```python
# print size to check
print(dec_ndays_smooth_all.shape)
print(dec_ndays_le_all.shape)
```

```python
# transpose so that it will plot properly
dec_ndays_smooth_all = dec_ndays_smooth_all.transpose()
dec_ndays_le_all = dec_ndays_le_all.transpose()
# print size to check
print(dec_ndays_smooth_all.shape)
print(dec_ndays_le_all.shape)
```

## Plot box and whisker plots

* Mask what goes into the box plot by significance
* Then plot box and whisker plots by season of the trends at each point

```python
def setBoxColor(boxplot, colors):

    # Set edge color of the outside and median lines of the boxes
    for element in ['boxes', 'medians']:
        for box, color in zip(boxplot[element], colors):
            plt.setp(box, color=color,linewidth=3)

    # Set the color of the whiskers and caps of the boxes
    for element in ['whiskers', 'caps']:
        for box, color in zip(
                zip(boxplot[element][::2], boxplot[element][1::2]), colors):
            plt.setp(box, color=color,linewidth=3)
```

```python
# find missing data and mask - otherwise boxplot fails
mask_ens_mean_le = ~np.isnan(dec_ndays_le_ens_mean)
mask_ens_mean_smooth = ~np.isnan(dec_ndays_smooth_ens_mean)

mask_all_le = ~np.isnan(dec_ndays_le_all)
mask_all_smooth = ~np.isnan(dec_ndays_smooth_all)
```

```python
#put the data into lists because the number of significant trends will be different by season and dataset
dec_ndays_ens_mean_le_masked = [d[m] for d, m in zip(dec_ndays_le_ens_mean.T,mask_ens_mean_le.T)]
dec_ndays_ens_mean_smooth_masked = [d[m] for d, m in zip(dec_ndays_smooth_ens_mean.T,mask_ens_mean_smooth.T)]

dec_ndays_all_le_masked = [d[m] for d, m in zip(dec_ndays_le_all.T,mask_all_le.T)]
dec_ndays_all_smooth_masked = [d[m] for d, m in zip(dec_ndays_smooth_all.T,mask_all_smooth.T)]
```

```python
# check the different lengths for seasons
print("number of points by decade for CESM2-LE ensemble mean")
print("2010s:"+str(len(dec_ndays_ens_mean_le_masked[0])))
print("2020s:"+str(len(dec_ndays_ens_mean_le_masked[1])))
print("2080s:"+str(len(dec_ndays_ens_mean_le_masked[6])))
print("2090s:"+str(len(dec_ndays_ens_mean_le_masked[7])))

print("number of points by decade for SMOOTH ensemble mean")
print("2010s:"+str(len(dec_ndays_ens_mean_smooth_masked[0])))
print("2020s:"+str(len(dec_ndays_ens_mean_smooth_masked[1])))
print("2080s:"+str(len(dec_ndays_ens_mean_smooth_masked[6])))
print("2090s:"+str(len(dec_ndays_ens_mean_smooth_masked[7])))

print("number of points by decade for CESM2-LE all")
print("2010s:"+str(len(dec_ndays_all_le_masked[0])))
print("2020s:"+str(len(dec_ndays_all_le_masked[1])))
print("2080s:"+str(len(dec_ndays_all_le_masked[6])))
print("2090s:"+str(len(dec_ndays_all_le_masked[7])))

print("number of points by decade for SMOOTH all")
print("2010s:"+str(len(dec_ndays_all_smooth_masked[0])))
print("2020s:"+str(len(dec_ndays_all_smooth_masked[1])))
print("2080s:"+str(len(dec_ndays_all_smooth_masked[6])))
print("2090s:"+str(len(dec_ndays_all_smooth_masked[7])))
```
