# Calculate Surface roughness from Rufmod experiments

* **Description**: Reads in and creates decadal averages from rufmod experiment
* **Input data**: Rufmod output in timeseries format
* **Output data**: Netcdf with processed ice state for all members
* **Creator**: Alice DuVivier
* **Date**: November 2021

The rufmod experiments were performed where the sea ice roughness over Arctic sea ice regions was set to be equal to what it would be over open ocean. This is to better understand ice-atmosphere coupling, processes, and feedbacks.

```python
# Get subsets for just short period at start (before ice loss is dramatic)
yy_st = 2015
yy_ed = 2020
yy = list(range(yy_st,yy_ed,1))
```

```python
# subset data for these years only, keep all months
sidragtop_le_sub = sidragtop_masked_le.isel(time=sidragtop_masked_le.time.dt.year.isin([yy]))
sidragtop_rufmod_sub = sidragtop_masked_rufmod.isel(time=sidragtop_masked_rufmod.time.dt.year.isin([yy]))
```

```python
le_temp = sidragtop_le_sub.isel(time=sidragtop_le_sub.time.dt.month.isin([1,2,3]))
```

```python
le_temp
```

```python
# Loop through seasons to get all the annual cycles - sidragtop

# make numpy array to fill and specify dimensions we want
avg_array_le = np.zeros([len(season_names),len(sidragtop_masked_le.nj),len(sidragtop_masked_le.ni)])
avg_array_rufmod = np.zeros([len(season_names),len(sidragtop_masked_rufmod.nj),len(sidragtop_masked_rufmod.ni)])
# for other method
avg_array_le_v2 = np.zeros([len(season_names),len(sidragtop_masked_le.nj),len(sidragtop_masked_le.ni)])
avg_array_rufmod_v2 = np.zeros([len(season_names),len(sidragtop_masked_rufmod.nj),len(sidragtop_masked_rufmod.ni)])


# season loop
#for ss in season_names[0:1]:
for ss in season_names:
    print(ss)
    if ss == 'JFM':
        count = 0
    else: 
        count = count+1
    #print(count)
    # get temporary array of just these months
    if ss == 'JFM':
        le_temp1 = sidragtop_le_sub.isel(time=sidragtop_le_sub.time.dt.month.isin([1,2,3]))
        rufmod_temp1 = sidragtop_rufmod_sub.isel(time=sidragtop_rufmod_sub.time.dt.month.isin([1,2,3]))
    if ss == 'AMJ':
        le_temp1 = sidragtop_le_sub.isel(time=sidragtop_le_sub.time.dt.month.isin([4,5,6]))
        rufmod_temp1 = sidragtop_rufmod_sub.isel(time=sidragtop_rufmod_sub.time.dt.month.isin([4,5,6]))
    if ss == 'JAS':
        le_temp1 = sidragtop_le_sub.isel(time=sidragtop_le_sub.time.dt.month.isin([7,8,9]))
        rufmod_temp1 = sidragtop_rufmod_sub.isel(time=sidragtop_rufmod_sub.time.dt.month.isin([7,8,9]))
    if ss == 'OND':
        le_temp1 = sidragtop_le_sub.isel(time=sidragtop_le_sub.time.dt.month.isin([10,11,12]))
        rufmod_temp1 = sidragtop_rufmod_sub.isel(time=sidragtop_rufmod_sub.time.dt.month.isin([10,11,12]))
        
    # now stack so we can average over both ensemble and time
    le_temp2 = le_temp1.stack(all_times=("time","member_id"))
    rufmod_temp2 = rufmod_temp1.stack(all_times=("time","member_id"))
    # calculate mean over all times
    le_temp3 = le_temp2.mean(dim='all_times')
    rufmod_temp3 = rufmod_temp2.mean(dim='all_times')
    # put into arrays
    avg_array_le[count,:,:] = le_temp3
    avg_array_rufmod[count,:,:] = rufmod_temp3
    
    # second version of calculations - average over time then ensemble
    le_temp2 = le_temp1.mean(dim="time").mean(dim="member_id")
    rufmod_temp2 = rufmod_temp1.mean(dim="time").mean(dim="member_id")
    # put into arrays
    avg_array_le_v2[count,:,:] = le_temp2
    avg_array_rufmod_v2[count,:,:] = rufmod_temp2
        
```

```python
# check sizes
print(avg_array_le.shape)
print(avg_array_rufmod.shape)
print(avg_array_le_v2.shape)
print(avg_array_rufmod_v2.shape)
```

```python
# convert the numpy array to a xarray
avg_le = xr.DataArray(avg_array_le,dims=('season','nj','ni'))
avg_rufmod = xr.DataArray(avg_array_rufmod,dims=('season','nj','ni'))
avg_le_v2 = xr.DataArray(avg_array_le_v2,dims=('season','nj','ni'))
avg_rufmod_v2 = xr.DataArray(avg_array_rufmod_v2,dims=('season','nj','ni'))
```

```python
# calculate difference
diff = avg_rufmod - avg_le
diff_v2 = avg_rufmod_v2 - avg_le_v2
```

```python
# add cyclic point
avg_le = gvutil.xr_add_cyclic_longitudes(avg_le,"ni")
avg_rufmod = gvutil.xr_add_cyclic_longitudes(avg_rufmod,"ni")
diff = gvutil.xr_add_cyclic_longitudes(diff,"ni")

avg_le_v2 = gvutil.xr_add_cyclic_longitudes(avg_le_v2,"ni")
avg_rufmod_v2 = gvutil.xr_add_cyclic_longitudes(avg_rufmod_v2,"ni")
diff_v2 = gvutil.xr_add_cyclic_longitudes(diff_v2,"ni")


TLAT = gvutil.xr_add_cyclic_longitudes(TLAT,"nlon")
TLON = gvutil.xr_add_cyclic_longitudes(TLON,"nlon")
```

```python
#Plot each season and difference
vmax_in = 30
vmin_in = 0
cmap_in = plt.cm.get_cmap('viridis')
vmax_diff = 10
vmin_diff = -10
cmap_diff = plt.cm.get_cmap('seismic')

#for ss in season_names[0:1]:
for ss in season_names:
    print('Plotting spatial means for month '+ss)
    if ss == 'JFM':
        count = 0
    else: 
        count = count+1
    #print(count)
    # create figure
    fig = plt.figure(figsize=(20,20))
    fout = 'cesm2_le_and_rufmod_'+var_in_1+'_difference_2015-2019_'+ss
    title = ss+' surface drag over sea ice - 2015-2019'
    
    # grab data we want to plot and add cyclic point
    temp_le = avg_le[count,:,:]
    temp_rufmod = avg_rufmod[count,:,:]
    temp_diff = diff[count,:,:]    

    # Make subplots - note it's nrow x ncol x index (starting upper left)
    # First subplot: CESM2-LE
    ax = fig.add_subplot(1,3,1, projection=ccrs.NorthPolarStereo())
    ax.add_feature(cfeature.LAND,zorder=100,edgecolor='k')
    ax.set_boundary(circle, transform=ax.transAxes)
    ax.set_extent([0.005, 360, 90, 55], crs=ccrs.PlateCarree())
    this=ax.pcolormesh(TLON,TLAT,
                       temp_le,
                       cmap=cmap_in,vmax=vmax_in,vmin=vmin_in,
                       transform=ccrs.PlateCarree())
    plt.colorbar(this,orientation='horizontal',label=units,fraction=0.04,pad=0.01)
    plt.title('CESM2-LE',fontsize=20)
    # Second subplot: Rufmod aka SMOOTH
    ax = fig.add_subplot(1,3,2, projection=ccrs.NorthPolarStereo())
    ax.add_feature(cfeature.LAND,zorder=100,edgecolor='k')
    ax.set_boundary(circle, transform=ax.transAxes)
    ax.set_extent([0.005, 360, 90, 55], crs=ccrs.PlateCarree())
    this=ax.pcolormesh(TLON,TLAT,
                       temp_rufmod,
                       cmap=cmap_in,vmax=vmax_in,vmin=vmin_in,
                       transform=ccrs.PlateCarree())
    plt.colorbar(this,orientation='horizontal',label=units,fraction=0.04,pad=0.01)
    plt.title('SMOOTH',fontsize=20)
    # First subplot: CESM2-LE
    ax = fig.add_subplot(1,3,3, projection=ccrs.NorthPolarStereo())
    ax.add_feature(cfeature.LAND,zorder=100,edgecolor='k')
    ax.set_boundary(circle, transform=ax.transAxes)
    ax.set_extent([0.005, 360, 90, 55], crs=ccrs.PlateCarree())
    this=ax.pcolormesh(TLON,TLAT,
                       temp_diff,
                       cmap=cmap_diff,vmax=vmax_diff,vmin=vmin_diff,
                       transform=ccrs.PlateCarree())
    plt.colorbar(this,orientation='horizontal',label=units,fraction=0.04,pad=0.01)
    plt.title('Difference(rufmod-cesm2le)',fontsize=20)
    
    # Finalize figure and save
    fig.suptitle(title,fontsize=15, y=0.75)  
    fig.subplots_adjust(bottom=0.45,wspace=0.1)
    #fig = plt.savefig(fout+'.png', bbox_inches='tight', dpi=200)    
```

```python
# Loop through seasons to get all the annual cycles - sidragtop

# make numpy array to fill and specify dimensions we want
avg_array_le = np.zeros([len(season_names),len(sidragtop_masked_le.nj),len(sidragtop_masked_le.ni)])
avg_array_rufmod = np.zeros([len(season_names),len(sidragtop_masked_rufmod.nj),len(sidragtop_masked_rufmod.ni)])
# for other method
avg_array_le_v2 = np.zeros([len(season_names),len(sidragtop_masked_le.nj),len(sidragtop_masked_le.ni)])
avg_array_rufmod_v2 = np.zeros([len(season_names),len(sidragtop_masked_rufmod.nj),len(sidragtop_masked_rufmod.ni)])


# season loop
#for ss in season_names[0:1]:
for ss in season_names:
    print(ss)
    if ss == 'JFM':
        count = 0
    else: 
        count = count+1
    #print(count)
    # get temporary array of just these months
    if ss == 'JFM':
        le_temp1 = sidragtop_le_sub.isel(time=sidragtop_le_sub.time.dt.month.isin([1,2,3]))
        rufmod_temp1 = sidragtop_rufmod_sub.isel(time=sidragtop_rufmod_sub.time.dt.month.isin([1,2,3]))
    if ss == 'AMJ':
        le_temp1 = sidragtop_le_sub.isel(time=sidragtop_le_sub.time.dt.month.isin([4,5,6]))
        rufmod_temp1 = sidragtop_rufmod_sub.isel(time=sidragtop_rufmod_sub.time.dt.month.isin([4,5,6]))
    if ss == 'JAS':
        le_temp1 = sidragtop_le_sub.isel(time=sidragtop_le_sub.time.dt.month.isin([7,8,9]))
        rufmod_temp1 = sidragtop_rufmod_sub.isel(time=sidragtop_rufmod_sub.time.dt.month.isin([7,8,9]))
    if ss == 'OND':
        le_temp1 = sidragtop_le_sub.isel(time=sidragtop_le_sub.time.dt.month.isin([10,11,12]))
        rufmod_temp1 = sidragtop_rufmod_sub.isel(time=sidragtop_rufmod_sub.time.dt.month.isin([10,11,12]))
        
    # now stack so we can average over both ensemble and time
    le_temp2 = le_temp1.stack(all_times=("time","member_id"))
    rufmod_temp2 = rufmod_temp1.stack(all_times=("time","member_id"))
    # calculate mean over all times
    le_temp3 = le_temp2.mean(dim='all_times')
    rufmod_temp3 = rufmod_temp2.mean(dim='all_times')
    # put into arrays
    avg_array_le[count,:,:] = le_temp3
    avg_array_rufmod[count,:,:] = rufmod_temp3
    
    # second version of calculations - average over time then ensemble
    le_temp2 = le_temp1.mean(dim="time").mean(dim="member_id")
    rufmod_temp2 = rufmod_temp1.mean(dim="time").mean(dim="member_id")
    # put into arrays
    avg_array_le_v2[count,:,:] = le_temp2
    avg_array_rufmod_v2[count,:,:] = rufmod_temp2
        
```

## Calculate the "open ocean" drag coefficient everywhere

The equation used is from CESM for open-ocean drag and common to many ocean models.
It is from Large and Pond 1981/1982.

```python
# all we *really* need is windspeed, so combine futures_3 and futures_4 to get just ws
ws = np.sqrt((uatm*uatm)+(vatm*vatm))
```

```python
# test that it worked properly
ws.sel(member_id=0).isel(time=0).plot()
```

```python
# set some constants for the calculation
# these are empirically derived
c1 = 0.0027
c2 = 0.000142
c3 = 0.0000764
```

```python
# calculate squareroot of the drag coefficient
rdn = np.sqrt((c1/ws) + c2 + (c3*ws))
```

```python
# calculate drag coefficient 
cdn = rdn*rdn
```

```python
# plot cdn to see if it looks reasonable
cdn.sel(member_id=0).isel(time=0).plot()
```
