## Read the CESM-LE data 

We will use [`intake-esm`](https://intake-esm.readthedocs.io/en/latest/), which is a data catalog tool.
It enables querying a database for the files we want, then loading those directly as an `xarray.Dataset`.

First step is to set the "collection" for the CESM-LE, which depends on a json file conforming to the [ESM Catalog Specification](https://github.com/NCAR/esm-collection-spec).

```python
catalog_file = '/glade/collections/cmip/catalog/intake-esm-datastore/catalogs/glade-cesm2-le.json'

cat = intake.open_esm_datastore(catalog_file)
```

```python
# list the variables to load
var_in = 'daidtd'

# aice, hi, sispeed, uvel, vvel, sidragtop, evap, frazil, congel, snoice, 
# meltt, meltb, meltl, dvidtd, dvidtt, daidtd, daidtt
```

Now we can use the [`to_dataset_dict`](https://intake-esm.readthedocs.io/en/latest/api.html#intake_esm.core.esm_datastore.to_dataset_dict) method to return a dictionary of `xarray.Dataset`'s. `intake_esm` makes groups of these according to rules in the collection spec file.

We can use the `preprocess` parameter to pass in a function that makes some corrections to the dataset. So first we define a function that does the following:
- fix the time coordinate to be the middle of the interval
- subset to the time-interval 1950-2100

```python
client
```

```python
%%time
with dask.config.set(**{'array.slicing.split_large_chunks': True}):
    dsets = subset.to_dataset_dict(cdf_kwargs={'chunks': {'time':240}, 'decode_times': True})

#dsets
```

```python
# print names of the dataset keys. 
dsets.keys()
```

```python
# Look at just one dataset to see what it looks like. 
# Note that for 1081 there is one member_id, but for 1231 there are 10 member_ids
# these refer to the individual ensemble members!

#dsets['ice.historical.cice.h.cmip6.1081.'+var_in]
#dsets['ice.historical.cice.h.cmip6.1231.'+var_in]
```

```python
# load in the historical and future datasets

historicals = []
futures = []

for key in sorted(dsets.keys()):
    if 'historical' in key:
        historicals.append(dsets[key])
        print(key)
    elif 'ssp370' in key:
        futures.append(dsets[key])
        print(key)
```

```python
# Now put these into an array
historical_ds = xr.concat(historicals, dim='member_id')
future_ds = xr.concat(futures, dim='member_id')
```

```python
# note that the historical and future xarray datasets have the same coordinates and dimensions *except* time, 
# so we need to concatenate over time
ds_all = xr.concat([historical_ds,future_ds],dim='time')
```

## Get decadal means

```python
#specify some arrays that will become coordinate arrays
decades = list(range(1950,2100,10))
months = list(range(1,13,1))

month_names = ['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun', 'Jul', 'Aug', 'Sep', 'Oct', 'Nov', 'Dec']
decade_names = ['1950','1960','1970','1980','1990','2000','2010','2020','2030',
                '2040','2050','2060','2070','2080','2090']
```

```python
# Loop through months and decades to get all the annual cycles

# make numpy array to fill and specify dimensions we want
avg_array = np.zeros([len(months),len(decades),len(ds_all.nj),len(ds_all.ni)])
var_array = np.zeros([len(months),len(decades),len(ds_all.nj),len(ds_all.ni)])
#decadal_mean.shape

# decade loop
for dec in decades:
    print('Calculating monthly means for '+str(dec)+'s')
    # need to specify which spot in the array we'll fill for the decade
    if dec == 1950:
        count = 0
    else: 
        count = count+1
    # set the start and end years for this decade
    yy_st = dec
    yy_ed = yy_st + 10
    yy = list(range(yy_st,yy_ed,1))
    # subset the data for these years only but keep all months
    temp = ds_all.isel(time=ds_all.time.dt.year.isin([yy]))

    # month loop
    for mm in months:
        #print('Finding month '+str(mm))
        # subset by month
        temp2 = temp.isel(time=temp.time.dt.month.isin([mm]))
        # stack so that we can average over both ensemble and time
        temp3 = temp2.stack(all_times=("time","member_id"))
        # calculate stats: mean & variance
        avg_array[(mm-1),count,:,:] = temp3.mean(dim='all_times')
        var_array[(mm-1),count,:,:] = temp3.var(dim='all_times')
```

```python
# check sizes
avg_array.shape
```

```python
# convert the numpy array to a xarray
decadal_avg = xr.DataArray(avg_array,dims=('month','decade','nj','ni'))
decadal_var = xr.DataArray(var_array,dims=('month','decade','nj','ni'))
```

```python
# assign some coordinate arrays
decadal_avg['decade'] = np.array(decades)
decadal_avg['month'] = np.array(months)
decadal_avg['nj'] = ds_all.nj
decadal_avg['ni'] = ds_all.ni

decadal_var['decade'] = np.array(decades)
decadal_var['month'] = np.array(months)
decadal_var['nj'] = ds_all.nj
decadal_var['ni'] = ds_all.ni
```
