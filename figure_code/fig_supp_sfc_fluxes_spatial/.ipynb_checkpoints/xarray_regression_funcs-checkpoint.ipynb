```python
import xarray as xr
import numpy as np
from scipy.stats import linregress,pearsonr, t
import matplotlib.pyplot as plt 
```

```python
time_len=20
y_len=50
x_len=60

spatial=xr.DataArray(np.random.randn(time_len,y_len,x_len),dims=['time','y','x'],coords={'time':range(time_len),'y':range(y_len),'x':range(x_len)})
tseries=xr.DataArray(np.random.randn(time_len),coords={'time':range(time_len)},dims=['time'])
```

```python
print(spatial)
print(tseries)
```

```python
def pvalue_array(x,y,dname):
    x_an=x-x.mean(dname)
    slope=((x_an)*(y-y.mean(dname))).sum(dname)/((x_an)*(x_an)).sum(dname)
    interc=y.mean(dname)-slope*x.mean(dname)
    ypred=slope*x+interc  
    n=len(x[dname])
    mse=np.sqrt(((y-ypred)**2).sum(dname)/(n-2))
    xsq=np.sqrt(((x_an)**2).sum(dname))
    standerr=mse/xsq
    pval=2*(1-t.cdf(np.abs(slope/standerr),n-2))
    pval=xr.DataArray(pval,dims=standerr.dims,coords=standerr.coords)
    return pval

def rvalue_array(x,y,dname):  
    xmean=x.mean(dname)
    ymean=y.mean(dname)
    numer=(x*y).mean(dname)-xmean*ymean
    denom=np.sqrt(((x**2).mean(dname)-xmean**2)*\
                  (((y**2).mean(dname))-(ymean**2)))
    return numer/denom

def regcoeff_array(x,y,dname):
    x_an=x-x.mean(dname)
    y_an=y-y.mean(dname)
    slope=(x_an*y_an).sum(dname)/(x_an*x_an).sum(dname)
    return slope
```

```python
%%time
regcoeff=regcoeff_array(tseries,spatial,'time')
rvalues=rvalue_array(tseries,spatial,'time')
pvalues=pvalue_array(tseries,spatial,'time')
```

```python
#double check that it matches the scipy methods
reg_objs=[[ linregress(tseries,spatial[:,yy,xx]) for yy in range(y_len)] for xx in range(x_len)]
sp_reg=np.array([[reg_objs[xx][yy].slope for yy in range(y_len)] for xx in range(x_len)])
sp_rvalue=np.array([[reg_objs[xx][yy].rvalue for yy in range(y_len)] for xx in range(x_len)])
scipy_rvalue=np.array([[pearsonr(tseries,spatial[:,yy,xx])[0] for yy in range(y_len)] for xx in range(x_len)])
sp_pvalue=np.array([[reg_objs[xx][yy].pvalue for yy in range(y_len)] for xx in range(x_len)])

f=plt.figure()
plt.pcolormesh(regcoeff.x,regcoeff.y,regcoeff.values-sp_reg.T)
plt.colorbar()

f=plt.figure()
plt.pcolormesh(rvalues.x,rvalues.y,rvalues.values-sp_rvalue.T)
plt.colorbar()

f=plt.figure()
plt.pcolormesh(rvalues.x,rvalues.y,rvalues.values-scipy_rvalue.T)
plt.colorbar()


f=plt.figure()
plt.pcolormesh(pvalues.x,pvalues.y,pvalues.values-sp_pvalue.T)
plt.colorbar()
```
