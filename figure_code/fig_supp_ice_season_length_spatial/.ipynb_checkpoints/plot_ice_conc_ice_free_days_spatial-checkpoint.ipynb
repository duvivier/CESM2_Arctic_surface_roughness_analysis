## Calculate average ice free days per decade

- Average by decade and over ensembles to get ensemble mean of ice covered days

```python
#specify some arrays that will become coordinate arrays
decades = list(range(2020,2100,10))
decade_names = ['2020','2030','2040','2050','2060','2070','2080','2090']
```

```python
# Loop through seasons - rufmod
print('Calculating Decadal and Ensemble Means ')

# make numpy array to fill and specify dimensions we want
dec_array_le = np.zeros([len(decade_names),len(data_le.nj),len(data_le.ni)])
dec_array_smooth = np.zeros([len(decade_names),len(data_smooth.nj),len(data_smooth.ni)])

# decade loop
for dec in decades:
    print('Calculating decadal means for '+str(dec)+'s')
    # need to specify which spot in the array we'll fill for the decade
    if dec == 2020:
        count = 0
    else: 
        count = count+1
    # set the start and end years for this decade
    yy_st = dec
    yy_ed = yy_st + 10
    yy = list(range(yy_st,yy_ed,1))
    # subset the data for these years only but keep all months
    temp1 = data_le.isel(time=data_le.time.isin([yy]))
    temp2 = data_smooth.isel(time=data_smooth.time.isin([yy]))
    # stack so we can average over years and ensembles at once
    temp1a = temp1.stack(all_times=("time","member_id"))
    dec_array_le[count,:,:] = temp1a.mean(dim='all_times')
    temp2a = temp2.stack(all_times=("time","member_id"))
    dec_array_smooth[count,:,:] = temp2a.mean(dim='all_times')        
```

```python
print(dec_array_smooth.shape)
print(dec_array_le.shape)
```

```python
# convert the numpy array to a xarray for easier plotting
dec_ndays_smooth = xr.DataArray(dec_array_smooth,dims=('decade','nj','ni'))
dec_ndays_le = xr.DataArray(dec_array_le,dims=('decade','nj','ni'))
```

```python
# set coordinate arrays
dec_ndays_smooth['decade'] = decade_names
dec_ndays_smooth['nj'] = data_smooth['nj'].values
dec_ndays_smooth['ni'] = data_smooth['ni'].values

dec_ndays_le['decade'] = decade_names
dec_ndays_le['nj'] = data_le['nj'].values
dec_ndays_le['ni'] = data_le['ni'].values
```

## Plotting

### General plotting data

```python
# make circular boundary for polar stereographic circular plots
theta = np.linspace(0, 2*np.pi, 100)
center, radius = [0.5, 0.5], 0.5
verts = np.vstack([np.sin(theta), np.cos(theta)]).T
circle = mpath.Path(verts * radius + center)
```

```python
# get pop grid for lat/lon values
grid = pop_tools.get_grid('POP_gx1v7')
# assign lat/lon data
TLONG = grid.TLONG
TLAT = grid.TLAT
```

```python
# Keep only NH points
TLONG_sub = TLONG.isel(nlat=slice(300,384))
TLAT_sub = TLAT.isel(nlat=slice(300,384))
```

```python
fig,ax=plt.subplots(1,3,figsize=(15,3))
axs=ax.flatten()
dec_ndays_le_sub.isel(decade=0).plot(ax=axs[0])
TLAT_sub.plot(ax=axs[1])
TLONG_sub.plot(ax=axs[2])
plt.subplots_adjust(wspace=0.4)
```

```python
# Setting up a target grid to only regrid data in the northern hemisphere north of ~50N
#Start at ~50N, but do not include 90N otherwise cartopy gets rage-y (i.e. it doesn't work) when plotting
lat=np.arange(49.4,90.0,0.5) 
lon=np.arange(0.0,361.0,1.0)
#create a meshgrid (2D fields of lats and lons)
lon2d,lat2d=np.meshgrid(lon,lat) 
#set up the target grid as an xarray Dataset
target_grid=xr.Dataset({'lat': (['y', 'x'], lat2d),'lon': (['y', 'x'], lon2d)})
```

```python
#input grid, output grid, method, keyword arguments
regridder=xe.Regridder(dec_ndays_le_sub, target_grid, 'nearest_s2d',periodic=True,reuse_weights=False)
```

```python
# actually do the regridding
dec_ndays_le_sub_regrid = regridder(dec_ndays_le_sub)
dec_ndays_smooth_sub_regrid = regridder(dec_ndays_smooth_sub)
```

```python
fig,ax=plt.subplots(1,2,figsize=(10,3))
axs=ax.flatten()
dec_ndays_le_sub.isel(decade=0).plot(ax=axs[0])
axs[0].set_title('Original')
dec_ndays_le_sub_regrid.isel(decade=0).plot(ax=axs[1])
axs[1].set_title('Regridded')
plt.subplots_adjust(wspace=0.3)
```

### Now Plot - spatial plots by season of the trends at each point and difference.

```python
#Plot each season and percent difference
levels_in = np.arange(0,370,10)
cmap_in = cmocean.cm.thermal
levels_diff = np.arange(-14,16,2)
cmap_diff = plt.cm.get_cmap('bwr')

units = 'days'
```

```python
## Plot decadal means
units = 'days'

# decade loop
for dec in decades:
    print('Calculating decadal means for '+str(dec)+'s')

    # set output name
    fout = 'ice_covered_days_'+str(dec)+'s'
    
    # grab data to plot
    temp_le = dec_ndays_le_sub_regrid.sel(decade=str(dec))
    temp_smooth = dec_ndays_smooth_sub_regrid.sel(decade=str(dec))
    diff = temp_smooth - temp_le

    # create figure
    fig = plt.figure(figsize=(20,20))
     
    # Make subplots - note it's nrow x ncol x index (starting upper left)
    ### First subplot: CESM2-LE
    ax = fig.add_subplot(1,3,1, projection=ccrs.NorthPolarStereo())
    ax.add_feature(cfeature.LAND,zorder=100,edgecolor='k')
    ax.set_boundary(circle, transform=ax.transAxes)
    ax.set_extent([0.005, 360, 90, 65], crs=ccrs.PlateCarree())
    # plot this month's trends
    this=ax.contourf(temp_le.lon,temp_le.lat,
                     temp_le,
                     cmap = cmap_in,                      
                     levels=levels_in,extend='max',
                     transform=ccrs.PlateCarree())
    # add colorbar
    cbar = plt.colorbar(this,orientation='horizontal',fraction=0.04,pad=0.01)
    cbar.ax.tick_params(labelsize=15, labelrotation=45)
    cbar.ax.set_xlabel(units, fontsize=15)
    plt.title('CESM2-LE',fontsize=20)
    
    ### Second subplot: CESM2-LE
    ax = fig.add_subplot(1,3,2, projection=ccrs.NorthPolarStereo())
    ax.add_feature(cfeature.LAND,zorder=100,edgecolor='k')
    ax.set_boundary(circle, transform=ax.transAxes)
    ax.set_extent([0.005, 360, 90, 65], crs=ccrs.PlateCarree())
    # plot this month's trends
    this=ax.contourf(temp_smooth.lon,temp_smooth.lat,
                     temp_smooth,
                     cmap = cmap_in,                      
                     levels=levels_in,extend='max',
                     transform=ccrs.PlateCarree())
    # add colorbar
    cbar = plt.colorbar(this,orientation='horizontal',fraction=0.04,pad=0.01)
    cbar.ax.tick_params(labelsize=15, labelrotation=45)
    cbar.ax.set_xlabel(units, fontsize=15)
    plt.title('SMOOTH',fontsize=20)

    ### Third subplot: Difference
    ax = fig.add_subplot(1,3,3, projection=ccrs.NorthPolarStereo())
    ax.add_feature(cfeature.LAND,zorder=100,edgecolor='k')
    ax.set_boundary(circle, transform=ax.transAxes)
    ax.set_extent([0.005, 360, 90, 65], crs=ccrs.PlateCarree())
    # plot this month's trends
    this=ax.contourf(diff.lon,diff.lat,
                     diff,
                     cmap = cmap_diff,                      
                     levels=levels_diff,extend='both',
                     transform=ccrs.PlateCarree())
    # add colorbar
    cbar = plt.colorbar(this,orientation='horizontal',fraction=0.04,pad=0.01)
    cbar.ax.tick_params(labelsize=15, labelrotation=45)
    cbar.ax.set_xlabel(units, fontsize=15)
    plt.title('Difference: SMOOTH - CESM2-LE',fontsize=20)

    # Finalize figure and save
    fig.suptitle(str(dec)+'s',fontsize=20, y=0.45)  
    #fig.subplots_adjust(bottom=0.45,wspace=0.1)
    fig = plt.savefig(fout+'.png', bbox_inches='tight', dpi=200)
    
```
