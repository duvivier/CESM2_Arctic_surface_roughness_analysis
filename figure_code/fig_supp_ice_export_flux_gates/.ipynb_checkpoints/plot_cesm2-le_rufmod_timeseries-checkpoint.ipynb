## Calculate seasonal means

```python
season_names = ['OND','JFM', 'AMJ', 'JAS','ANN']
```

```python
# find total years
xarr_le = fram_le.coords['time.year'][(fram_le.coords['time.month']==1)]
xarr_smooth = fram_smooth.coords['time.year'][(fram_smooth.coords['time.month']==1)]
```

```python
fram_le.isel(time=fram_le.time.dt.month.isin([1,2,3]))
```

```python
# Loop through seasons - rufmod
print('Calculating Seasonal Means - SMOOTH')

# make numpy array to fill and specify dimensions we want
fram_seas_array_smooth = np.zeros([len(fram_smooth.member_id),len(season_names),len(xarr_smooth)])
barents_seas_array_smooth = np.zeros([len(fram_smooth.member_id),len(season_names),len(xarr_smooth)])
bering_seas_array_smooth = np.zeros([len(fram_smooth.member_id),len(season_names),len(xarr_smooth)])

for ss in season_names:
    print(ss)
    if ss == 'OND':
        s_count = 0
    else: 
        s_count = s_count+1
    # get temporary array of just these month by season
    if ss == 'JFM':
        temp0 = fram_smooth.isel(time=fram_smooth.time.dt.month.isin([1,2,3]))
        temp1 = barents_smooth.isel(time=barents_smooth.time.dt.month.isin([1,2,3]))
        temp2 = bering_smooth.isel(time=bering_smooth.time.dt.month.isin([1,2,3]))
    if ss == 'AMJ':
        temp0 = fram_smooth.isel(time=fram_smooth.time.dt.month.isin([4,5,6]))
        temp1 = barents_smooth.isel(time=barents_smooth.time.dt.month.isin([4,5,6]))
        temp2 = bering_smooth.isel(time=bering_smooth.time.dt.month.isin([4,5,6]))
    if ss == 'JAS':
        temp0 = fram_smooth.isel(time=fram_smooth.time.dt.month.isin([7,8,9]))
        temp1 = barents_smooth.isel(time=barents_smooth.time.dt.month.isin([7,8,9]))
        temp2 = bering_smooth.isel(time=bering_smooth.time.dt.month.isin([7,8,9]))
    if ss == 'OND':
        temp0 = fram_smooth.isel(time=fram_smooth.time.dt.month.isin([10,11,12]))
        temp1 = barents_smooth.isel(time=barents_smooth.time.dt.month.isin([10,11,12]))
        temp2 = bering_smooth.isel(time=bering_smooth.time.dt.month.isin([10,11,12]))
    if ss == 'ANN':
        temp0 = fram_smooth
        temp1 = barents_smooth
        temp2 = bering_smooth

    # now loop through years to get the seasonal average by year for each ensemble member
    for yy in xarr_smooth:
        if yy == 2015:
            y_count = 0
        else: 
            y_count = y_count+1 
        # select only the indexes for this year
        temp0a = temp0.isel(time=temp0.time.dt.year.isin([yy]))
        fram_seas_array_smooth[:,s_count,y_count] = temp0a.mean(dim='time')
        temp1a = temp1.isel(time=temp1.time.dt.year.isin([yy]))
        barents_seas_array_smooth[:,s_count,y_count] = temp1a.mean(dim='time')
        temp2a = temp2.isel(time=temp2.time.dt.year.isin([yy]))
        bering_seas_array_smooth[:,s_count,y_count] = temp2a.mean(dim='time')
        
```

```python
# Loop through seasons - CESM2-LE
print('Calculating Seasonal Means - CESM2-LE')

# make numpy array to fill and specify dimensions we want
fram_seas_array_le = np.zeros([len(fram_le.member_id),len(season_names),len(xarr_le)])
barents_seas_array_le = np.zeros([len(fram_le.member_id),len(season_names),len(xarr_le)])
bering_seas_array_le = np.zeros([len(fram_le.member_id),len(season_names),len(xarr_le)])

for ss in season_names:
    print(ss)
    if ss == 'OND':
        s_count = 0
    else: 
        s_count = s_count+1
    # get temporary array of just these month by season
    if ss == 'JFM':
        temp0 = fram_le.isel(time=fram_le.time.dt.month.isin([1,2,3]))
        temp1 = barents_le.isel(time=barents_le.time.dt.month.isin([1,2,3]))
        temp2 = bering_le.isel(time=bering_le.time.dt.month.isin([1,2,3]))
    if ss == 'AMJ':
        temp0 = fram_le.isel(time=fram_le.time.dt.month.isin([4,5,6]))
        temp1 = barents_le.isel(time=barents_le.time.dt.month.isin([4,5,6]))
        temp2 = bering_le.isel(time=bering_le.time.dt.month.isin([4,5,6]))
    if ss == 'JAS':
        temp0 = fram_le.isel(time=fram_le.time.dt.month.isin([7,8,9]))
        temp1 = barents_le.isel(time=barents_le.time.dt.month.isin([7,8,9]))
        temp2 = bering_le.isel(time=bering_le.time.dt.month.isin([7,8,9]))
    if ss == 'OND':
        temp0 = fram_le.isel(time=fram_le.time.dt.month.isin([10,11,12]))
        temp1 = barents_le.isel(time=barents_le.time.dt.month.isin([10,11,12]))
        temp2 = bering_le.isel(time=bering_le.time.dt.month.isin([10,11,12]))
    if ss == 'ANN':
        temp0 = fram_le
        temp1 = barents_le
        temp2 = bering_le

    # now loop through years to get the seasonal average by year for each ensemble member
    for yy in xarr_le:
        if yy == 2015:
            y_count = 0
        else: 
            y_count = y_count+1 
        # select only the indexes for this year
        temp0a = temp0.isel(time=temp0.time.dt.year.isin([yy]))
        fram_seas_array_le[:,s_count,y_count] = temp0a.mean(dim='time')
        temp1a = temp1.isel(time=temp1.time.dt.year.isin([yy]))
        barents_seas_array_le[:,s_count,y_count] = temp1a.mean(dim='time')
        temp2a = temp2.isel(time=temp2.time.dt.year.isin([yy]))
        bering_seas_array_le[:,s_count,y_count] = temp2a.mean(dim='time')
        
```

```python
print(fram_seas_array_le.shape)
print(fram_seas_array_smooth.shape)
```

```python
# convert the numpy array to a xarray
fram_seas_le = xr.DataArray(fram_seas_array_le,dims=('member_id','season','time'))
fram_seas_smooth = xr.DataArray(fram_seas_array_smooth,dims=('member_id','season','time'))

barents_seas_le = xr.DataArray(barents_seas_array_le,dims=('member_id','season','time'))
barents_seas_smooth = xr.DataArray(barents_seas_array_smooth,dims=('member_id','season','time'))

bering_seas_le = xr.DataArray(bering_seas_array_le,dims=('member_id','season','time'))
bering_seas_smooth = xr.DataArray(bering_seas_array_smooth,dims=('member_id','season','time'))
```

```python
# set coordinate arrays
fram_seas_le['season'] = season_names
fram_seas_le['time'] = xarr_le
fram_seas_le['member_id'] = fram_le['member_id']

fram_seas_smooth['season'] = season_names
fram_seas_smooth['time'] = xarr_smooth
fram_seas_smooth['member_id'] = fram_smooth['member_id']

barents_seas_le['season'] = season_names
barents_seas_le['time'] = xarr_le
barents_seas_le['member_id'] = fram_le['member_id']

barents_seas_smooth['season'] = season_names
barents_seas_smooth['time'] = xarr_smooth
barents_seas_smooth['member_id'] = fram_smooth['member_id']

bering_seas_le['season'] = season_names
bering_seas_le['time'] = xarr_le
bering_seas_le['member_id'] = fram_le['member_id']

bering_seas_smooth['season'] = season_names
bering_seas_smooth['time'] = xarr_smooth
bering_seas_smooth['member_id'] = fram_smooth['member_id']
```

## Calculate ensemble means & significance

```python
# calculate ensemble means
fram_seas_ens_mean_le = fram_seas_le.mean(dim='member_id')
fram_seas_ens_mean_smooth = fram_seas_smooth.mean(dim='member_id')

barents_seas_ens_mean_le = barents_seas_le.mean(dim='member_id')
barents_seas_ens_mean_smooth = barents_seas_smooth.mean(dim='member_id')

bering_seas_ens_mean_le = bering_seas_le.mean(dim='member_id')
bering_seas_ens_mean_smooth = bering_seas_smooth.mean(dim='member_id')
```

```python
# calculate statistical significance
sigval = 0.05

# calculate statistical significance
tval,pval = ttest_ind(fram_seas_le,fram_seas_smooth,axis=0,equal_var=False)
fram_seas_ens_mean_le_sig = fram_seas_ens_mean_le.where(pval <= sigval)
fram_seas_ens_mean_smooth_sig = fram_seas_ens_mean_smooth.where(pval <= sigval)

tval,pval = ttest_ind(barents_seas_le,barents_seas_smooth,axis=0,equal_var=False)
barents_seas_ens_mean_le_sig = barents_seas_ens_mean_le.where(pval <= sigval)
barents_seas_ens_mean_smooth_sig = barents_seas_ens_mean_smooth.where(pval <= sigval)

tval,pval = ttest_ind(bering_seas_le,bering_seas_smooth,axis=0,equal_var=False)
bering_seas_ens_mean_le_sig = bering_seas_ens_mean_le.where(pval <= sigval)
bering_seas_ens_mean_smooth_sig = bering_seas_ens_mean_smooth.where(pval <= sigval)
```

```python
# bootstrap the CESM2-LE 1000 times
n_bs=1000
fram_seas_bs_array_le = np.zeros([len(season_names),len(xarr_le),n_bs])
barents_seas_bs_array_le = np.zeros([len(season_names),len(xarr_le),n_bs])
bering_seas_bs_array_le = np.zeros([len(season_names),len(xarr_le),n_bs])

for ii in range(n_bs):
    # choose random ensemble member numbers
    mem1=np.random.choice(range(0,49))
    mem2=np.random.choice(range(0,49))
    mem3=np.random.choice(range(0,49))
    mem4=np.random.choice(range(0,49))
    mem5=np.random.choice(range(0,49)) 
    # now average those random ensembles
    temp = np.stack((fram_seas_le.isel(member_id=mem1),fram_seas_le.isel(member_id=mem2),fram_seas_le.isel(member_id=mem3),
                     fram_seas_le.isel(member_id=mem4),fram_seas_le.isel(member_id=mem5)),axis=-1)
    fram_seas_bs_array_le[:,:,ii]=np.mean(temp,axis=2)
    temp = np.stack((barents_seas_le.isel(member_id=mem1),barents_seas_le.isel(member_id=mem2),barents_seas_le.isel(member_id=mem3),
                     barents_seas_le.isel(member_id=mem4),barents_seas_le.isel(member_id=mem5)),axis=-1)
    barents_seas_bs_array_le[:,:,ii]=np.mean(temp,axis=2)
    temp = np.stack((bering_seas_le.isel(member_id=mem1),bering_seas_le.isel(member_id=mem2),bering_seas_le.isel(member_id=mem3),
                     bering_seas_le.isel(member_id=mem4),bering_seas_le.isel(member_id=mem5)),axis=-1)
    bering_seas_bs_array_le[:,:,ii]=np.mean(temp,axis=2)
    
```

```python
fram_seas_bs_array_le.shape
```

```python
# convert the numpy array to a xarray for easier plotting
fram_seas_bs_le = xr.DataArray(fram_seas_bs_array_le,dims=('season','time','bootstrap_num'))
barents_seas_bs_le = xr.DataArray(barents_seas_bs_array_le,dims=('season','time','bootstrap_num'))
bering_seas_bs_le = xr.DataArray(bering_seas_bs_array_le,dims=('season','time','bootstrap_num'))
```

```python
# assign coordinate arrays
fram_seas_bs_le['season'] = fram_seas_le.season
fram_seas_bs_le['time']= fram_seas_le.time

barents_seas_bs_le['season'] = fram_seas_le.season
barents_seas_bs_le['time']= fram_seas_le.time

bering_seas_bs_le['season'] = fram_seas_le.season
bering_seas_bs_le['time']= fram_seas_le.time
```
